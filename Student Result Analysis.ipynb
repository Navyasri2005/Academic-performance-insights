```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed Column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Gender Distribution

```python
palette = {'male': 'orange', 'female': 'blue'}
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender", hue="Gender", palette=palette, legend=False)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Gender Distribution")
plt.show()
```

# From the above chart we have analysed that: 
The number of females in the data is more than the number of males 

# Relationship Between Parent's Education and Student's Score

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship Between Parent's Education and Student's Score")
plt.show()
```

# From The Above Chart We Have Concluded That:
The Education Of The Parents Have a Good Impact On Their Scores

# Relationship Between Parent's Martial Status and Student's Score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship Between Parent's Martial Status and Student's Score")
plt.show()
```

# From The Above Chart We Have Concluded That:
There is no/negligible impact on the student's score due to their parent's marital status

# Distribution Of Ethnic Groups

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[df['EthnicGroup'] == "group A"].count()
groupB = df.loc[df['EthnicGroup'] == "group B"].count()
groupC = df.loc[df['EthnicGroup'] == "group C"].count()
groupD = df.loc[df['EthnicGroup'] == "group D"].count()
groupE = df.loc[df['EthnicGroup'] == "group E"].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

print(mlist)
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title("Distribution Of Ethnic Groups")
plt.show()

```

```python
palette = {
    'group A': 'orange',
    'group B': 'blue',
    'group C': 'green',
    'group D': 'red',
    'group E': 'purple'
}

plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='EthnicGroup', hue='EthnicGroup', palette=palette, legend=False)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```
